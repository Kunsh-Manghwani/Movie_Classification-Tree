```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Movie_classificationother.csv",header=0)
```

```python
df.head()
```

```python
df.info()
```

## Missing value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(),inplace=True)
```

```python
df.info()
```

## Dummy variable creation

```python
df.head()
```

```python
df = pd.get_dummies(df,columns =["Genre","3D_available"],drop_first=True )
```

```python
df.head()
```

## X-Y split

```python
x = df.loc[:,df.columns!="Start_Tech_Oscar"]
type(x)
```

```python
x.head()
```

```python
y = df['Start_Tech_Oscar']
type(y)
```

```python
y.head()
```

```python
y.shape
```

```python
x.shape
```

## Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2,random_state =0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

## Training Classification Tree

```python
from sklearn import tree
```

```python
clftree = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clftree.fit(x_train,y_train)
```

## Predict values using trained models

```python
y_train_pred = clftree.predict(x_train)
y_test_pred = clftree.predict(x_test)
```

```python
y_test_pred
```

## Model performance

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_train,y_train_pred)
```

### here 172 and 92 are our right readings 

```python
confusion_matrix(y_test,y_test_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
bag_clf = BaggingClassifier(base_estimator=cltree,n_estimators=1000,bootstrap=True,n_jobs=-1,random_state=42)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(n_estimators=1000,n_jobs=-1,random_state =42)
```

```python
rf_clf.fit(x_train,y_train)
```

```python
confusion_matrix(y_test,rf_clf.predict(x_test))
```

```python
accuracy_score(y_test,rf_clf.predict(x_test))
```

## Grid search

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf = RandomForestClassifier(random_state=42)
```

```python
params_grid = {"max_features" :[4,5,6,7,8,9,10],
               "min_samples_split" : [2,3,10]}
```

```python
grid_search = GridSearchCV(rf_clf,params_grid,n_jobs=-1,cv=5,scoring="accuracy")
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
cvrf_clf = grid_search.best_estimator_
```

```python
accuracy_score(y_test,cvrf_clf.predict(x_test))
```

```python
confusion_matrix(y_test,cvrf_clf.predict(x_test))
```

## XG boost

```python
import xgboost as xgb
```

### we can deffrentiate parametres into 3:
General Parameters:- that guide the overall funcctioning
booster parameters:- guide the indivisual boosting tree/regression at each step
Learning task Parameters:- guide the optimization Performed

```python
xg_clf = xgb.XGBClassifier(max_depth = 5 , n_estimators = 10000 , learning_rate =0.3, n_jobs = -1)
```

```python
xg_clf.fit(x_train,y_train)
```

```python
accuracy_score(y_test,xg_clf.predict(x_test))
```

```python
xgb.plot_importance(xg_clf,color = "brown")
```

## This is importance graph

```python
xgb_clf = xgb.XGBClassifier(n_estimators = 250 , learning_rate =0.1, n_jobs = -1,random_state = 42)
```

```python
param1 = {
    'max_depth':range(3,10,2),
    'gamma':[0.1,0.2,0.3],
    'subsample':[0.8,0.9],
    'colsample_bytree':[0.8,0.9],
    'reg_alpha':[1e-2,0.1,1]
}
```

```python
grid_search = GridSearchCV(xgb_clf, param1,n_jobs=-1,cv=5,scoring='accuracy')
```

```python
grid_search.fit(x_train,y_train)
```

```python
cvxg_clf = grid_search.best_estimator_
```

```python
accuracy_score(y_test,cvxg_clf.predict(x_test))
```

```python
grid_search.best_params_
```
